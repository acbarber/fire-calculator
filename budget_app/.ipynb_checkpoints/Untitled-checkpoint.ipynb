```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import locale
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' )
```

```python
a=[10000]
for i in range(30):
    a.append(a[-1]*1.03)
sns.set()
ax = sns.lineplot(range(31),a,markers='o')
ax.set_title("What happens when you don't pay off your debt")
ax.set_xlabel('Years')
ax.set_ylabel('Debt')
xt = [0,10000,12000,14000,16000,18000,20000,22000,24000]
ax.set_yticklabels([locale.currency(i, grouping=True) for i in xt])
ax.text(3+22,24200,f'{str(locale.currency(a[-1],grouping=True))} ->',horizontalalignment='center',verticalalignment='center')
ax.get_figure().savefig('debt.jpg')
```
